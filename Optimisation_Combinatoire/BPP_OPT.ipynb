# **First-Fit-Decreasing**

```python
def heuristic_FFD(c,w):
  n = len(w)
  order = sorted([i for i in range(n)],lambda i:w[i],reversed=True)
  bin_for_item = [-1 for i in range(n)]
  bin_space = []
  for i in order:
    for j in range(len(bin_space)):
      if w[i]<bin_space[j]:
        bin_for_item[i]=j
        bin_space[j]-=w[i]
        break
    if bin_for_item[i] < 0:
      j = len(bin_space)
      bin_for_item[i] = j
      bin_space.append(c-w[i])
  n_bin = len(bin_space)
  return n_bin, bin_for_item
```

# **First-Fit-Increasing**

```python
def heuristic_FFI(c,w):
  n = len(w)
  order = sorted([i for i in range(n)],lambda i:w[i])
  bin_for_item = [-1 for i in range(n)]
  bin_space = []
  for i in order:
    for j in range(len(bin_space)):
      if w[i]<bin_space[j]:
        bin_for_item[i]=j
        bin_space[j]-=w[i]
        break
    if bin_for_item[i] < 0:
      j = len(bin_space)
      bin_for_item[i] = j
      bin_space.append(c-w[i])
  n_bin = len(bin_space)
  return n_bin, bin_for_item
```

# **Best-Fit**

```python
def heuristic_BF(c,w):
  n = len(w)
  bin_for_item = [-1 for i in range(n)]
  bin_space = []
  for i in order:
    tmp = sorted(bin_space)
    k=0
    while True:
      #k = bin_space.index(min(bin_space))
      if w[i]<tmp[k]:
        j = bin_space.index(tmp[k])
        bin_for_item[i]=j
        bin_space[j]-=w[i]
        break
      else:
        k+=1
      
    if bin_for_item[i] < 0:
      j = len(bin_space)
      bin_for_item[i] = j
      bin_space.append(c-w[i])
  n_bin = len(bin_space)
  return n_bin, bin_for_item
```

# **Worst-Fit**

```python
def heuristic_WF(c,w):
  n = len(w)
  bin_for_item = [-1 for i in range(n)]
  bin_space = []
  for i in order:
    k = bin_space.index(max(bin_space))
    if w[i]<bin_space[k]:
      bin_for_item[i]=k
      bin_space[k]-=w[i]
      
    if bin_for_item[i] < 0:
      j = len(bin_space)
      bin_for_item[i] = j
      bin_space.append(c-w[i])
  n_bin = len(bin_space)
  return n_bin, bin_for_item
```
